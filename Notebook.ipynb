```python
X = data[['number of bedrooms', 'number of bathrooms','living area','condition of the house', 'Number of schools nearby',]]
y = data['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    'criterion': ['mse', 'friedman_mse', 'mae'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = tree_model, param_grid = param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_predictions = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_predictions)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
linear_predictions = linear_model.predict(X_test)
```

```python
mean_absolute_error(y_test, linear_predictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
```

```python
param_grid_rf = {
    'max_depth': [5, 10, 15],
    'n_estimators': [2, 3, 4, 5, 6, 7, 8, 9, 10],
}
```

```python
grid_rfr = GridSearchCV(estimator = rf_model, param_grid = param_grid_rf)
```

```python
grid_rfr.fit(X_train, y_train)
```

```python
grid_rfr.best_params_
```

```python
rfr_predictions = grid_rfr.predict(X_test)
```

```python
mean_absolute_error(y_test, rfr_predictions)
```

```python
sample = [[3, 2, 1500, 3, 1]]  # bedrooms, bathrooms, area, condition, schools nearby
prediction = grid_rfr.predict(sample)
print("Predicted price:", prediction)
```
